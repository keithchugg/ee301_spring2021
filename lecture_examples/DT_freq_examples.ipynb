# Discrete Time Sinusoid Examples (DT frequency examples)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nu = [ 0, 0.25, 0.5, 0.75, 1] 
n = np.arange(20)

for nu0 in nu:
    plt.figure()
    plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
    plt.legend()
```

```python
#####################################################################################
### exploring DT frequency via plots
#####################################################################################
n = np.arange(20)
t = np.arange(0, 20, 0.001)
nu = [ 0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.9 ] 
for nu0 in nu:
    plt.figure()
    plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
    plt.plot(t, np.cos(2 * np.pi * nu0 * t), c='g', linewidth=0.75, label='cont. time')
    plt.legend()

```

```python
#####################################################################################
### exploring DT frequency via plots
#####################################################################################
n = np.arange(20)
t = np.arange(0, 20, 0.001)
nu0 = 2. * np.pi / 3.0  
print(f'nu = {nu0}')

plt.figure()
plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
plt.plot(t, np.cos(2 * np.pi * nu0 * t), c='g', linewidth=0.75, label='cont. time')

plt.legend()

print(f'cos[2 pi nu n]: {np.cos(2 * np.pi * nu0 * n)}')
```
