# Discrete Time Sinusoid Examples (DT frequency examples)

```python
import numpy as np
import matplotlib.pyplot as plt
```

Plot some examples of $\cos( 2 \pi \nu_0 n)$ for a list of $\nu_0$ values.

```python
nu = [ 0, 0.25, 0.5, 0.75, 1] 
n = np.arange(20)

for nu0 in nu:
    plt.figure()
    plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
    plt.legend()
```

Notice from above that $\cos( 2 \pi (1/4) n ) = \cos( 2 \pi (3/4) n )$.  Let's plot $\cos(2\pi \nu_0t)$ for "continuous" time $t$ to see what is going on.

```python
#####################################################################################
### exploring DT frequency via plots
#####################################################################################
n = np.arange(20)
t = np.arange(0, 20, 0.001)
nu = [ 0, 0.25, 0.5, 0.75, 1] 
for nu0 in nu:
    plt.figure()
    plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
    plt.plot(t, np.cos(2 * np.pi * nu0 * t), c='g', linewidth=0.75, label='cont. time')
    plt.legend()

```

So what is happening is that for larger values of $\nu_0$, the continuous time signal is going through multiple cycles before it is "observed" again at integer times.  This is an example of what is known as frequency aliasing.  Here are a couple of links to youtube to see this effect:

* [Helicoptor flies with rotars that are still??](https://www.youtube.com/watch?v=jWIcVP6GRfw)
* Sometimes people call this the [wagon wheel effect](https://www.youtube.com/watch?v=6XwgbHjRo30)
* Here's an [experiment designed to show this effect](https://www.youtube.com/watch?v=QOwzkND_ooU).  

All of these are caused by the finite frame-rate of the camera.  For example, if the video is at 30 frames/sec, in the helicopter example the rotars are making an integer number of rotations between video frames. Specifically, if the rotars were rotating at 4 revolutions every 30 msec, this is what we would see.  

This can happen with spatial frequencies as well -- i.e., when the spatial resolution of a camera aliases high-frequency spatial patterns (i.e., small checkers/plaid).  This is called a [Moire pattern in photography](https://www.youtube.com/watch?v=ZFteoCptzvM). This was a common effect observed with older, low-res TVs when people wore checked patterns.  




Let's next get some additional insight into why $\nu_0$ has to be rational in order for $\cos( 2 \pi \nu_0 n)$ to be periodic.  Notice below that the signal is not peridoic -- we can view it as smapling a CT periodic signal, but out of synch so that the samples never repeat.  

A similar phenomenon is that if you consider $x(t) = \cos(2\pi f_0 t) +  \cos(2\pi f_1 t)$ this signal is not perdiodic.  Can you determine the relation between $f_0$ and $f_1$ that must hold for this to be periodic?

```python
#####################################################################################
### exploring DT frequency via plots
#####################################################################################
n = np.arange(20)
t = np.arange(0, 20, 0.001)
nu0 = 2. * np.pi / 3.0  # nu is not a rational number, Omega is not a rational multiple of 2 pi
print(f'nu = {nu0}')

plt.figure()
plt.stem(n, np.cos(2 * np.pi * nu0 * n), label=f'DT: nu = {nu0}')
plt.plot(t, np.cos(2 * np.pi * nu0 * t), c='g', linewidth=0.75, label='cont. time')

plt.legend()

print(f'cos[2 pi nu n]: {np.cos(2 * np.pi * nu0 * n)}')
```

Finally, we saw that $\cos( 2 \pi (1/4) n) = \cos( 2 \pi (3/4) n)$, but below we see that $\sin( 2 \pi (1/4) n) \neq \sin( 2 \pi (3/4) n)$.  You should see where this signal moves on the unit circle to undersatnd why this is the case.

```python
n = np.arange(20)

nu0 = 0.25
nu1 = 0.75

plt.figure()
plt.stem(n, np.cos(2 * np.pi * nu0 * n), 'b', label=f'DT: cos: nu = {nu0}')
plt.stem(n, np.cos(2 * np.pi * nu1 * n), 'g', label=f'DT: cos: nu = {nu1}')
plt.legend()

plt.figure()
plt.stem(n, np.sin(2 * np.pi * nu0 * n), 'b', label=f'DT: sin: nu = {nu0}')
plt.stem(n, np.sin(2 * np.pi * nu1 * n), 'g', label=f'DT: sin: nu = {nu1}')
plt.legend()
```
