## Sampling Theorem Workbook

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_samples = 10
# generate some random samples
x_samples = np.random.normal(0, 10, N_samples)

t = np.arange(-3, N_samples + 3, 0.001)
t_samples = np.arange(N_samples)
```

```python
x_hat = np.zeros(len(t))
sinc_terms = np.zeros((N_samples, len(t)))
for n in range(N_samples):
    sinc_terms[n] = x_samples[n] * np.sinc(t - n)
    x_hat += sinc_terms[n]
```

```python
plt.figure(figsize=(12, 8))
for n in range(N_samples):
    plt.plot(t, sinc_terms[n], linestyle='--')
plt.stem(t_samples, x_samples)
plt.xlabel('t')
plt.ylabel('sinc interpolation terms')
```

```python
plt.figure(figsize=(12, 8))
for n in range(N_samples):
    plt.plot(t, sinc_terms[n], linestyle='--')
plt.stem(t_samples, x_samples, label='samples')
plt.plot(t, x_hat, c='k', label='reconstruction')
plt.xlabel('t')
plt.ylabel('sinc interpolated samples')
```
