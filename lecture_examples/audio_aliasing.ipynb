## Aliasing Example Using Digital Audio

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
import os
import soundfile as sf
```

```python
wav_name = 'data/chugg_welcome.wav'
wav_name = 'data/chirp.wav'
audio, fs = sf.read(wav_name)  # fs = 48 kHz is assumed.

t_audio = np.arange(len(audio)) / fs
```

```python
plt.figure()
plt.plot(t_audio, audio, color = 'b')
#plt.legend()
plt.grid(linestyle=':')
t_half = t_audio[ int( len(t_audio) / 2) ]
# plt.xlim([t_half - 0.01, t_half + 0.01])
plt.xlabel("time (sec)")
#plt.ylabel("signals")


```

```python
## downsample by 4 to go fs=48 kHz to fs = 12 kHz

ds = 4

audio_alias = audio[::ds]
fd = int(fs / ds)
td = np.arange(len(audio_alias)) / ds


plt.figure()
plt.plot(td, audio_alias, color = 'r')
#plt.legend()
plt.grid(linestyle=':')
t_half = td[ int( len(td) / 2) ]
#plt.xlim([t_half - 0.01, t_half + 0.01])
plt.xlabel("time (sec)")
#plt.ylabel("signals")

base_wav = os.path.basename(wav_name).split('.')[0]

sf.write(f'data/{base_wav}_{fd/1000}_khz_aliasing.wav', audio_alias, fd)
```
